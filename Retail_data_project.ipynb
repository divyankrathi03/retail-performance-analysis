```python
import pandas as pd
```

```python
df = pd.read_excel('/content/retail_raw_data.xlsx')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Units_Sold'].fillna(1,inplace=True)
```

```python
df['Price'].fillna(55000,inplace=True)
```

```python
df['Discount'].fillna(0, inplace = True)
```

```python
df
```

```python
df['Net_Price'] = df['Price'] * (1-df['Discount'])
```

```python
df['Revenue'] = df['Net_Price'] * df['Units_Sold']
```

```python
df
```

```python
print("Total REvenue is ",df['Revenue'].sum())

print("Total units sold by all store ",df['Units_Sold'].sum())
```

```python
Revenue_Store = df.groupby('Store')['Revenue'].sum().sort_values(ascending=False).reset_index()
print(Revenue_Store)
```

```python
Revenue_Region = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False).reset_index()
print(Revenue_Region)
```

```python
Revenue_Product = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False).reset_index()
print(Revenue_Product)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x = 'Product', y = 'Revenue' , data=Revenue_Product )
plt.grid()
plt.show()
```

```python
sns.barplot(x = 'Region', y= 'Revenue' , data=Revenue_Region )
plt.grid()
plt.show()
```

**Key Findings**

Revenue is highly concentrated in West and South Region , contributing approximately 57% of total revenue.

East region underperforms primarily due to lower units sold, not pricing.

Phones (A + B) account for a majority of total revenue, creating concentration risk.

Accessories generate high volume but significantly lower revenue compared to electronics.

Heavy discounting on  reduces charger net revenue without proportional volume increase.

**Business Recommendations**

Increase marketing focus or distribution efforts in East region.

Reduce dependency on phone products by promoting accessories or laptops.

Test controlled price increases on high-volume, low-revenue products like chargers.

Review discount strategy for products with poor discount-to-volume response.
